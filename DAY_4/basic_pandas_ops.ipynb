```python
DATA_PATH: str = r"D:\CodeBase\Personal\RustyPanda\data\titanic_train.csv"
```

```python
df = pd.read_csv(
    DATA_PATH,
    dtype={
        'PassengerId': np.uint16,
        'Survived': bool,
        'Pclass': 'category',
        'Sex': 'category',
        'Age': np.float16,
        'SibSp': np.uint8,
        'Parch': np.uint8,
        'Fare': np.float16,
        'Embarked': 'category'
    },
    # chunksize=100
)
```

```python
df
```

```python
df.groupby('Sex')[['Fare', 'Age']].agg([min, max, np.mean, np.median, np.std, len])
```

```python
df.groupby('Sex')[['Survived']].agg(['count', sum, lambda x: x.sum()/x.count()])
```

```python
df.groupby('Sex').agg(
    total_count = ('Survived', 'count'),
    survived_count = ('Survived', 'sum'),
    percent_survived = ('Survived', lambda x: x.sum()*100/x.count()),
)
```
